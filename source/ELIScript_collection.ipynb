### Retrieving links 

```python
def retrieve_links():
    """ This function collects links from e-LIS site and appends it to a list.
    
    Parameters
    ----------
        url : str 
            string of a URL's site assigned to a variable,  ex: url= "http://site.com/"
    
    Returns
    -------
        final_url : list of str.
            list containing strings of URL.
    """   

    url="http://eprints.rclis.org/view/countries/"
    page = urllib.request.urlopen(url)
    soup = BeautifulSoup(page, 'html.parser')

    get_link = []
    

    for link in soup.find_all('a'):
        get_link.append(link.get('href'))
        
    return get_link

retrieved_links = retrieve_links()
```

```python
## todos os links da pagina, as proximas coisas s√£o pra limpar e pegar somente os paises em seus continentes
retrieved_links
```

```python
## separa os links de cada pais de acordo com seu continente
def get_list(separators, mylist):

    result = []
    sublist = []

    for x in mylist:
        if x in separators:
            if sublist:
                result.append(sublist)
            sublist = []
        else:
            sublist.append(x)

    result.append(sublist)
    
    return result


## faz um dicionario, no qual as chaves sao os continentes e os valores sao os paises
def get_dict(separators, mylist):
    
    mydict = {}
    
    result = get_list(separators, mylist)
    
    separators.remove('AQ-.html')
    
    
    mydict['ANTARCTICA.html'] = ['AQ-.html']
      
    
    for sep, res in zip(separators, result):
            mydict[sep] = res
    
    return mydict
```

```python
cont = ["AFRICA", "NCA", "SouthAmerica", "AQ-", "ASIA", "EUROPE", "OCEANIA"]

# pegar somente os links de paises
countries = [t for t in retrieved_links if ".html" in t and 'http' not in t and 'countries' not in t 
             and 'Melanesia' not in t]

# adiciona ".html" no final do nome dos continentes, pra bater com o nome na lista retrieved_links
separators = [c + '.html' for c in cont]
mydict = get_dict(separators, countries)

mydict
```

```python
def get_url_countries(values):
    url = "http://eprints.rclis.org/view/countries/"  
    return [url + v for v in values]

## criando a url total dos paises
```

### Defining Mozilla Firefox Profile

```python
def firefox_profile(path_to_save):

    profile = webdriver.FirefoxProfile()
    profile.set_preference("browser.preferences.instantApply",True)
    profile.set_preference("browser.helperApps.neverAsk.saveToDisk", "text/plain, application/octet-stream, application/binary, text/csv, application/csv, application/excel, text/comma-separated-values, text/xml, application/xml")
    profile.set_preference("browser.helperApps.alwaysAsk.force",False)
    profile.set_preference("browser.download.manager.showWhenStarting",False)
    profile.set_preference("browser.download.folderList", 2)
    profile.set_preference("browser.download.dir", path_to_save)

    driver = webdriver.Firefox(firefox_profile=profile)
    
    return driver
```

### Creating a loop to access every link and download the data available

```python
from selenium.common.exceptions import NoSuchElementException

def download_eLIS_data(final_url, driver):

    for country in final_url:
        driver.get(country)

        try:
            select = Select(driver.find_element_by_name("format"))                              
            select.select_by_value("CSV")
            export_path = "//input[@name='_action_export_redir'][@type='submit']"
            elem = driver.find_element_by_xpath(export_path).click()
            time.sleep(60)
        
        except NoSuchElementException:  # spelling error making this code not work as expected
            print("Cannot download. Please, download it manually here: http://eprints.rclis.org/view/countries/ES-.html")
            continue
        except TimeoutException:
            raise TimeoutError("Your request has been timed out! Took too long to load connections link")
    
    driver.quit()
```

```python
for k, v in mydict.items():
    
    path_continent = os.path.join(wdir, k[:-5])  ## esse :-5 √© para tirar os '.html' do nome do continente
    
    # create folder continent
    create_folder(path_continent)
    
    # retrieve
    final_url = get_url_countries(v)
    driver = firefox_profile(path_continent)
    download_eLIS_data(final_url, driver)
    
### pegar o dicionario mydict[continente] = [lista de paises] e vai rodar a coleta para cada continente, salvando
# em sua pastinha
```
